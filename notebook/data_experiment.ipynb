## Lets Visualize the data

```python
import matplotlib.pyplot as plt
```

The relationship between Study hour and marks is good one for linear regression

number of courses vs marks not a linear one

```python
plt.plot(data['time_study'], data['Marks'], c='b', label='Study Hour vs Marks')
plt.plot(data['number_courses'], data['Marks'], c='r', label='Number of Course vs Marks')
plt.legend()
plt.title('Student Marks Analysis')
plt.xlabel('Features')
plt.ylabel('Marks')
plt.savefig('images/image1.png')
plt.show();
```

```python
plt.scatter(data['time_study'], data['Marks'], c='b', label='Study Hour vs Marks')
plt.scatter(data['number_courses'], data['Marks'], c='r', label='Number of Course vs Marks')
plt.legend()
plt.title('Student Marks Analysis')
plt.xlabel('Features')
plt.ylabel('Marks')
plt.savefig('images/image2.png')
plt.show();
```

```python
hour_vs_marks = data.drop('number_courses', axis=1)
df = pd.DataFrame(hour_vs_marks)
```

```python
df.head()
```

```python
plt.plot(df['time_study'], df['Marks'])
plt.title('Study_hours vs Marks')
plt.xlabel("Hours")
plt.ylabel('Marks')
plt.show();
```

```python
plt.scatter(df['time_study'], df['Marks'])
plt.title('Study_hours vs Marks')
plt.xlabel("Hours")
plt.ylabel('Marks')
plt.savefig('images/image3.png')
plt.show();
```
